```python
import pandas as pd
```

Dataset description 

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations 
2. Temp : temperature noon (temperature max)  in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90 
4. Ws :Wind speed in km/h: 6 to 29 
5. Rain: total day in mm: 0 to 16.8
FWI Components  
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5 
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9 
8. Drought Code (DC) index from the FWI system:  7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5 
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely   Fire and No fire

```python
# import dataset 
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv' , header = 1)
df.head()
```

```python
# data check 
#df.info()

# checking null values 
print('Null values in the dataset')
df[df.isnull().any(axis=1)]   

# null values as a percentage 
#print('Null values as a percentage')
#df.isnull().sum()/df.shape[0]*100 
```

```python

# We have 2 region in this dataset - Sidi-Bel Abbes and Bejaia Region. Adding regions as a column 


# adding the region flag 

df.loc[122:,'is_sidi_bel_region'] = 1
df.loc[:122,'is_sidi_bel_region'] = 0

df['is_sidi_bel_region'].value_counts()

# converting the region flag to numerical 
df['is_sidi_bel_region'] = df['is_sidi_bel_region'].astype(int) 

# null value treatment (Since only 2 "any" null value rows drop those rows)
df_1 = df.dropna().reset_index(drop=True)

# checking null values
print("",df_1.isnull().sum())

# checking the dataset 
display("data snapshot" , df_1.head())

# column names  
print("column names \n " , df_1.columns)

# there is a extra space in the couple of column name. Correcting it

print(type(df_1.columns))

df_1.columns = df_1.columns.str.strip()


# checking column names  
print("formatted column names \n " , df_1.columns)

```

```python
# checking the column data types

#print("column data types \n" , df_1.dtypes)

# removing a row with a numeric day type 

df_1 = df_1[df_1['day'] != 'day']


# converting few columns to numerical

df_1[['day','month','year','Temperature','RH','Ws']] = df_1[['day','month','year','Temperature','RH','Ws']].astype(int)

# dataframe datatypes 

#print("Recheck column data types \n" , df_1.dtypes)

# convert the remaining columns to float 

# finding the columns that are not int

not_int_columns =    [i  for i in df_1.columns if df_1[i].dtype != 'int64']

print("Columns that are not int \n" , not_int_columns)

# checking the "class" datatype

print(df_1['Classes'].value_counts())

# remove classes from the not_int_columns list (because it can be converted to float)

not_int_columns.remove('Classes')

for i in not_int_columns:
    df_1[i] = df_1[i].astype(float) 

```

```python
# final dataset 

display("Final dataset \n" , df_1.head())

# final datatypes 

display("Final datatypes \n" , df_1.dtypes)
```

```python
# dump the data 

df_1.to_csv('Algerian_forest_fires_dataset_UPDATE_CLEAN.csv')
```
